```python
x = np.array([[2],[4],[5],[6],[7],[9]])
y = np.array([30000,50000,70000,85000,100000,120000])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
coff = lr.coef_[0]
coff
```

```python
intercept = lr.intercept_
intercept
```

```python
n = int(input("enter the experience"))
pred = coff*n+intercept
```

```python
import numpy as np
```

```python
x = np.array([2,4,5,6,7,9])
y = np.array([30000,50000,70000,85000,100000,120000])
x
```

```python
mymodel = np.poly1d(np.polyfit(x,y,4))
```

```python
pred3 = mymodel(x)
pred3
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='red',marker='+')
plt.plot(x,pred3)
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y,pred3))
```
